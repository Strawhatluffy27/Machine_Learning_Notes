```python
import pandas as pd
```

# reading XML File using Python

```python
pd.read_xml("flower.xml",xpath="flower")
```

```python
pd.read_xml("flower.xml",xpath="farmer")
```

```python
pd.read_xml("flower.xml",xpath="country")
```

```python
data =pd.read_xml("flower.xml",xpath="flower")
```

```python
data
```

```python
data.to_xml("flower1.xml")
```

```python
data.to_xml("flower12.xml",attr_cols=["sl","sw","pl","label"])
```

```python
data.to_xml("flower123.xml",elem_cols=["sl","sw","pl","label"])
```
